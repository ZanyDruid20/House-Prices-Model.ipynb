```python
predictors = ["interest", "vacancy", "adjusted_price","adjusted_value"]
target = "change"
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
# this function is going to have a test set, a predictor and a target
# at the end of the function we return a predictor
START = 260 # weeks that add up to five years
STEP =  52 # how many weeks are in a year
def predict(train,test,predictors,target):
  # this code prevents overfitting
  rf = RandomForestClassifier(min_samples_split=10, random_state=1)
  rf.fit(train[predictors],train[target])
  preds = rf.predict(test[predictors])
  return preds
def backtest(data,predictors,target):
  all_preds = []
  for i in range(START,data.shape[0],STEP):
    # we split the data into train and test
    # we split the data up to the year
    # then we traverse up to other years
    train = price_data.iloc[:i]
    test = price_data.iloc[i:(i+STEP)]
    all_preds.append(predict(train,test,predictors,target))
  preds = np.concatenate(all_preds)
  return preds, accuracy_score(data.iloc[START:][target],preds)
```

```python
preds,accuracy = backtest(price_data,predictors,target)
```

```python
accuracy
```

```python
yearly = price_data.rolling(52,min_periods=1).mean()
```

```python
yearly
```

```python
yearly_ratios = [p + "_year" for p in predictors]
price_data[yearly_ratios] = price_data[predictors] / yearly[predictors]
```

```python
price_data
```

```python
preds,accuracy = backtest(price_data,predictors + yearly_ratios, target)
```

```python
accuracy
```

```python
pred_match = (preds == price_data.iloc[START:][target])
pred_match[pred_match == True] = "blue"
pred_match[pred_match == False] = "red"
```

```python
import matplotlib.pyplot as plt

plot_data = price_data.iloc[START:].copy()
# reset index data frame is turned into index
plot_data.reset_index().plot.scatter(x="index", y="adjusted_price", color=pred_match)
```

```python
# this code checks for which column is the most important and could help you to
# to drop or add to make it a better model
from sklearn.inspection import permutation_importance
rf = RandomForestClassifier(min_samples_split=10, random_state=1)
# we fit our model accross every data
rf.fit(price_data[predictors],price_data[target])
result = permutation_importance(rf,price_data[predictors],price_data[target],n_repeats=10,random_state=1)
```

```python
result.importances_mean # this tells us whuch columns were more important to our model
# and it tells which one was more factored into the predictions
# and the most important one was adjusted_value
```
